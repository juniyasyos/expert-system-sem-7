## 2ï¸âƒ£ Knowledge Base - Basis Pengetahuan

**Knowledge Base** adalah tempat menyimpan pengetahuan dalam sistem pakar. Terdiri dari:
- **Fakta (Facts)**: Informasi yang diketahui benar
- **Aturan (Rules)**: Hubungan logis IF-THEN

### Implementasi Knowledge Base Class

```python
class KnowledgeBase:
    """
    Knowledge Base untuk menyimpan fakta dan aturan
    """
    
    def __init__(self, verbose=True):
        self.facts = set()  # Set untuk menyimpan fakta (unique)
        self.rules = []     # List untuk menyimpan aturan
        self.verbose = verbose  # Kontrol output
    
    def add_fact(self, fact: str, show_message=None):
        """Menambahkan fakta ke knowledge base"""
        self.facts.add(fact)
        show = show_message if show_message is not None else self.verbose
        if show:
            print(f"  âœ“ Fakta: '{fact}'")
    
    def add_rule(self, name: str, conditions: List[str], conclusion: str, 
                 operator: str = "AND", show_message=None):
        """
        Menambahkan aturan ke knowledge base
        
        Args:
            name: Nama aturan (misal: "R1")
            conditions: List kondisi (IF part)
            conclusion: Kesimpulan (THEN part)
            operator: "AND" atau "OR"
            show_message: Override verbose untuk rule ini
        """
        rule = {
            'name': name,
            'conditions': conditions,
            'conclusion': conclusion,
            'operator': operator
        }
        self.rules.append(rule)
        show = show_message if show_message is not None else self.verbose
        if show:
            cond_str = f" {operator} ".join(conditions)
            print(f"  âœ“ {name}: IF {cond_str} THEN {conclusion}")
    
    def has_fact(self, fact: str) -> bool:
        """Cek apakah fakta ada di knowledge base"""
        return fact in self.facts
    
    def display_facts(self, title="FAKTA YANG DIKETAHUI"):
        """Tampilkan semua fakta"""
        print("\n" + "="*70)
        print(title)
        print("="*70)
        if not self.facts:
            print("  (tidak ada fakta)")
        else:
            for i, fact in enumerate(sorted(self.facts), 1):
                print(f"  {i:2d}. {fact}")
        print()
    
    def display_rules(self, title="ATURAN YANG TERSEDIA"):
        """Tampilkan semua aturan"""
        print("\n" + "="*70)
        print(title)
        print("="*70)
        if not self.rules:
            print("  (tidak ada aturan)")
        else:
            for i, rule in enumerate(self.rules, 1):
                conditions_str = f" {rule['operator']} ".join(rule['conditions'])
                print(f"\n  [{i}] {rule['name']}:")
                print(f"      IF   {conditions_str}")
                print(f"      THEN {rule['conclusion']}")
        print()

# Test Knowledge Base
print("Knowledge Base class berhasil dibuat! âœ“")
```

## 3ï¸âƒ£ Inference Engine - Mesin Inferensi

**Inference Engine** adalah komponen yang melakukan reasoning (penalaran) dengan menggunakan fakta dan aturan yang ada di knowledge base.

### Forward Chaining Inference Engine
**Forward Chaining** adalah metode inferensi yang dimulai dari fakta yang diketahui, lalu menerapkan aturan untuk menghasilkan fakta baru sampai mencapai kesimpulan.

**Algoritma:**
1. Mulai dengan fakta awal
2. Cari aturan yang kondisinya terpenuhi
3. Eksekusi aturan â†’ tambahkan kesimpulan sebagai fakta baru
4. Ulangi sampai tidak ada aturan yang bisa dieksekusi

```python
class ForwardChainingEngine:
    """
    Forward Chaining Inference Engine
    Data-driven reasoning: Dari fakta â†’ kesimpulan
    """
    
    def __init__(self, knowledge_base: KnowledgeBase):
        self.kb = knowledge_base
        self.inference_chain = []  # Untuk tracking proses inferensi
    
    def can_fire(self, rule: dict) -> bool:
        """
        Cek apakah aturan bisa dieksekusi (fired)
        """
        conditions = rule['conditions']
        operator = rule['operator']
        
        if operator == "AND":
            # Semua kondisi harus terpenuhi
            return all(cond in self.kb.facts for cond in conditions)
        elif operator == "OR":
            # Minimal satu kondisi terpenuhi
            return any(cond in self.kb.facts for cond in conditions)
        else:
            return False
    
    def fire_rule(self, rule: dict, verbose: bool = True) -> bool:
        """
        Eksekusi aturan: tambahkan kesimpulan ke fakta
        Returns True jika berhasil menambahkan fakta baru
        """
        conclusion = rule['conclusion']
        
        if conclusion not in self.kb.facts:
            self.kb.facts.add(conclusion)
            self.inference_chain.append({
                'rule': rule['name'],
                'conclusion': conclusion
            })
            if verbose:
                print(f"  ğŸ”¥ {rule['name']} FIRED â†’ '{conclusion}'")
            return True
        return False
    
    def run(self, max_iterations: int = 100, verbose: bool = True, 
            show_iterations: bool = True, show_final_summary: bool = True):
        """
        Jalankan forward chaining
        
        Args:
            max_iterations: Maksimal iterasi (default 100)
            verbose: Tampilkan proses detail
            show_iterations: Tampilkan info per iterasi
            show_final_summary: Tampilkan ringkasan akhir
        """
        if verbose:
            print("\n" + "="*70)
            print("âš™ï¸  FORWARD CHAINING - Proses Inferensi Dimulai")
            print("="*70)
        
        iteration = 0
        self.inference_chain = []
        total_fired = 0
        
        while iteration < max_iterations:
            iteration += 1
            fired_count = 0
            
            if verbose and show_iterations:
                print(f"\nâ”Œâ”€ Iterasi {iteration} " + "â”€" * 55)
            
            # Coba fire setiap aturan
            for rule in self.kb.rules:
                if self.can_fire(rule):
                    if self.fire_rule(rule, verbose=verbose):
                        fired_count += 1
                        total_fired += 1
            
            if verbose and show_iterations:
                if fired_count == 0:
                    print("  â””â”€ Tidak ada aturan yang dieksekusi")
                else:
                    print(f"  â””â”€ {fired_count} aturan berhasil dieksekusi")
            
            # Jika tidak ada aturan yang fired, berhenti
            if fired_count == 0:
                if verbose:
                    print("\n" + "â”€"*70)
                    print("âœ“ Inferensi selesai - tidak ada aturan lagi yang dapat dieksekusi")
                break
        
        if verbose and show_final_summary:
            print("\n" + "="*70)
            print("ğŸ“Š RINGKASAN INFERENSI")
            print("="*70)
            print(f"  â€¢ Total iterasi         : {iteration}")
            print(f"  â€¢ Total aturan dieksekusi: {total_fired}")
            print(f"  â€¢ Fakta baru dihasilkan : {len(self.inference_chain)}")
            print(f"  â€¢ Total fakta akhir     : {len(self.kb.facts)}")
            print("="*70)
        
        return self.inference_chain
    
    def display_results(self, show_facts=True, show_inference_chain=True):
        """
        Tampilkan hasil inferensi
        
        Args:
            show_facts: Tampilkan semua fakta akhir
            show_inference_chain: Tampilkan rantai inferensi
        """
        if show_inference_chain:
            print("\n" + "="*70)
            print("ğŸ”— RANTAI INFERENSI (Fakta Baru yang Diturunkan)")
            print("="*70)
            if not self.inference_chain:
                print("  (tidak ada fakta baru yang diturunkan)")
            else:
                for i, step in enumerate(self.inference_chain, 1):
                    print(f"  {i:2d}. '{step['conclusion']}'")
                    print(f"      â””â”€ dari aturan: {step['rule']}")
            print()
        
        if show_facts:
            self.kb.display_facts(title="ğŸ“š SEMUA FAKTA AKHIR (Awal + Hasil Inferensi)")

# Test Forward Chaining Engine
print("Forward Chaining Engine berhasil dibuat! âœ“")
```

## 4ï¸âƒ£ Backward Chaining - Goal-Driven Reasoning

**Backward Chaining** adalah metode inferensi yang dimulai dari tujuan (goal) yang ingin dibuktikan, lalu bekerja mundur untuk mencari fakta pendukung.

**Algoritma:**
1. Mulai dengan goal yang ingin dibuktikan
2. Cari aturan yang kesimpulannya = goal
3. Cek apakah semua kondisi aturan tersebut terpenuhi
4. Jika kondisi belum terpenuhi, jadikan sebagai sub-goal dan ulangi
5. Jika semua kondisi terpenuhi, goal terbukti

```python
class BackwardChainingEngine:
    """
    Backward Chaining Inference Engine
    Goal-driven reasoning: Dari tujuan â†’ fakta pendukung
    """
    
    def __init__(self, knowledge_base: KnowledgeBase):
        self.kb = knowledge_base
        self.proof_tree = []  # Untuk tracking proses pembuktian
        self.depth = 0  # Untuk indentasi output
    
    def prove(self, goal: str, verbose: bool = True) -> bool:
        """
        Buktikan apakah goal dapat dicapai
        
        Args:
            goal: Tujuan yang ingin dibuktikan
            verbose: Print proses pembuktian
        
        Returns:
            True jika goal terbukti, False jika tidak
        """
        indent = "  " * self.depth
        
        if verbose:
            print(f"{indent}ğŸ¯ Mencoba membuktikan: '{goal}'")
        
        # Base case: goal sudah ada di fakta
        if goal in self.kb.facts:
            if verbose:
                print(f"{indent}  âœ“ Sudah ada di fakta!")
            return True
        
        # Cari aturan yang kesimpulannya = goal
        rules_for_goal = [r for r in self.kb.rules if r['conclusion'] == goal]
        
        if not rules_for_goal:
            if verbose:
                print(f"{indent}  âœ— Tidak ada aturan untuk '{goal}'")
            return False
        
        # Coba buktikan dengan setiap aturan
        for rule in rules_for_goal:
            if verbose:
                print(f"{indent}  Menggunakan {rule['name']}")
            
            # Cek semua kondisi
            self.depth += 1
            conditions_met = True
            
            for condition in rule['conditions']:
                if not self.prove(condition, verbose):
                    conditions_met = False
                    break
            
            self.depth -= 1
            
            # Jika semua kondisi terpenuhi
            if conditions_met:
                if verbose:
                    print(f"{indent}  âœ“ '{goal}' terbukti!")
                
                # Tambahkan ke fakta
                self.kb.facts.add(goal)
                self.proof_tree.append({
                    'goal': goal,
                    'rule': rule['name'],
                    'conditions': rule['conditions']
                })
                return True
        
        if verbose:
            print(f"{indent}  âœ— '{goal}' tidak dapat dibuktikan")
        return False
    
    def run(self, goal: str, verbose: bool = True) -> bool:
        """
        Jalankan backward chaining untuk membuktikan goal
        """
        if verbose:
            print("\n" + "="*60)
            print("BACKWARD CHAINING - Proses Pembuktian")
            print("="*60)
            print(f"\nGoal: '{goal}'")
            print("-"*60 + "\n")
        
        self.proof_tree = []
        self.depth = 0
        result = self.prove(goal, verbose)
        
        if verbose:
            print("\n" + "="*60)
            if result:
                print(f"âœ“ BERHASIL: Goal '{goal}' terbukti!")
            else:
                print(f"âœ— GAGAL: Goal '{goal}' tidak dapat dibuktikan")
            print("="*60)
        
        return result
    
    def display_proof_tree(self):
        """Tampilkan pohon pembuktian"""
        if not self.proof_tree:
            print("\nTidak ada pohon pembuktian")
            return
        
        print("\n" + "="*60)
        print("POHON PEMBUKTIAN")
        print("="*60)
        for step in self.proof_tree:
            print(f"\nGoal: '{step['goal']}'")
            print(f"  Aturan: {step['rule']}")
            print(f"  Kondisi yang dipenuhi:")
            for cond in step['conditions']:
                print(f"    â€¢ {cond}")

# Test Backward Chaining Engine
print("Backward Chaining Engine berhasil dibuat! âœ“")
```

## 6ï¸âƒ£ Visualisasi Decision Tree (Sederhana)

Mari kita visualisasi proses reasoning dengan ASCII art untuk memahami alur keputusan.

```python
def visualize_inference_chain(inference_chain):
    """
    Visualisasi proses inferensi dengan ASCII tree
    """
    print("\n" + "="*60)
    print("ğŸŒ³ DECISION TREE - Alur Inferensi")
    print("="*60)
    
    if not inference_chain:
        print("\nTidak ada inferensi yang dilakukan")
        return
    
    print("\nFakta Awal")
    print("    â†“")
    
    for i, step in enumerate(inference_chain, 1):
        rule = step['rule']
        conclusion = step['conclusion']
        
        print(f"    [{rule}]")
        print(f"    â†“")
        print(f"    '{conclusion}'")
        
        if i < len(inference_chain):
            print("    â†“")
    
    print("\nâœ“ Kesimpulan Akhir Tercapai")
    print("="*60)

# Visualisasi untuk contoh diagnosis
visualize_inference_chain(medical_engine.inference_chain)
```

### Yang Telah Dipelajari:

1. **Knowledge Base** 
   - Representasi fakta dan aturan
   - Struktur IF-THEN rules
   
2. **Forward Chaining**
   - Data-driven reasoning
   - Dari fakta awal â†’ kesimpulan
   - Cocok untuk: diagnosis, klasifikasi, sistem rekomendasi

3. **Backward Chaining**
   - Goal-driven reasoning  
   - Dari tujuan â†’ cari fakta pendukung
   - Cocok untuk: pembuktian, verifikasi, troubleshooting

### Perbedaan Forward vs Backward Chaining:

| Aspek | Forward Chaining | Backward Chaining |
|-------|------------------|-------------------|
| **Arah** | Fakta â†’ Kesimpulan | Goal â†’ Fakta |
| **Strategi** | Data-driven | Goal-driven |
| **Mulai dari** | Fakta yang diketahui | Tujuan yang ingin dibuktikan |
| **Cocok untuk** | Discovery, diagnosis | Verification, planning |
| **Efisiensi** | Bisa generate banyak fakta tidak relevan | Fokus hanya pada goal |
